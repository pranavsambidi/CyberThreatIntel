```python
import os
import pymupdf4llm
from langchain_text_splitters import RecursiveCharacterTextSplitter

print("Environment verified. Libraries loaded successfully.")
```

```python
# Load a single sample PDF to prove the ingestion works
sample_pdf = "../data/APT29 attacks Embassies using CVE-2023-38831 - report en.pdf" # Make sure this matches a real file name
md_text = pymupdf4llm.to_markdown(sample_pdf)

# Prove chunking works
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = text_splitter.split_text(md_text)

print(f"Successfully loaded PDF.")
print(f"Total Markdown Length: {len(md_text)} characters")
print(f"Sliced into {len(chunks)} chunks for Vectorization.")
print("\nSample Chunk Preview:")
print("-" * 40)
print(chunks[0][:300] + "...")
```
